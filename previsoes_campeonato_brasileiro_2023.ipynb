```python
tabela_jogos_ajustada = tabela_jogos.set_index('Casa \ Fora')
tabela_jogos_ajustada = tabela_jogos_ajustada.unstack().reset_index()
tabela_jogos_ajustada = tabela_jogos_ajustada.rename(columns={'level_0': 'fora',
                                                             'Casa \ Fora': 'casa',
                                                             0: 'resultado'})
tabela_jogos_ajustada
```

```python
def ajustar_apelido_time(linha):
    return dic_times[linha["fora"]]
```

```python
tabela_jogos_ajustada['fora'] = tabela_jogos_ajustada.apply(ajustar_apelido_time, axis=1)
tabela_jogos_ajustada
```

```python
tabela_jogos_ajustada = tabela_jogos_ajustada.query('fora != casa')
tabela_jogos_ajustada
```

```python
tabela_jogos_ajustada['resultado'] = tabela_jogos_ajustada['resultado'].fillna('a jogar')
tabela_jogos_ajustada.isnull().sum()
```

```python
tabela_jogos_realizados = tabela_jogos_ajustada[tabela_jogos_ajustada['resultado'].str.contains('–')]
tabela_jogos_realizados= tabela_jogos_realizados[['casa', 'fora', 'resultado']]
tabela_jogos_realizados
```

```python
#usamos o sinal '~' quando queremos a condição contrária. Nesse caso, o que NÃO contém o '-'
tabela_jogos_faltantes = tabela_jogos_ajustada[~tabela_jogos_ajustada['resultado'].str.contains('–')]
tabela_jogos_faltantes = tabela_jogos_faltantes.drop('resultado', axis=1).reset_index(drop=True)
tabela_jogos_faltantes = tabela_jogos_faltantes[['casa', 'fora']]
tabela_jogos_faltantes
```

```python
# o parametro Expand é para aplicar o resultado dos splits em colunas diferentes, ao invés de retornar uma lista
tabela_jogos_realizados[['gols_feitos_casa', 'gols_sofridos_casa']] = (tabela_jogos_realizados['resultado']
                                                                       .str.split('–', expand=True)
                                                                      .astype('int'))
                                                                    
```

```python
media_gols_casa = tabela_jogos_realizados.groupby('casa').mean(numeric_only=True)
media_gols_fora = (tabela_jogos_realizados.groupby('fora').mean(numeric_only=True)
                   .rename(columns={"gols_feitos_casa": "gols_sofridos_fora", 
                          'gols_sofridos_casa': 'gols_feitos_fora'}))
```

```python
#juntamos as tabelas e mantemos os indices das 2
tabela_stats = (media_gols_casa
                .merge(media_gols_fora, right_index=True, left_index=True)
                .reset_index().rename(columns={'casa': 'times'})
                )

```

```python
brasileirao = '''


30ª rodada – 2º turno
25/10 – Flamengo 1 x 0 Red Bull Bragantino
25/10 – Botafogo 0 x 1 Cuiabá
25/10 – Palmeiras 1 x 0 Bahia
25/10 – Corinthians 1 x 1 Santos
25/10 – Goiás 1 x 1 Vasco
25/10 – Atlético-MG 2 x 0 Fluminense
25/10 – Internacional 3 x 4 Coritiba
25/10 – Athletico-PR 1 x 1 São Paulo
25/10 – Fortaleza 0 x 1 Cruzeiro
25/10 – América-MG 3 x 4 Grêmio
31ª rodada – 2º turno
29/10 – Flamengo 1 x 2 Santos
29/10 – Botafogo 3 x 4 Palmeiras
29/10 – São Paulo 1 x 0 Cruzeiro
29/10 – Corinthians 1 x 0 Athletico-PR
29/10 – Goiás 0 x 2 Red Bull Bragantino
29/10 – Atlético-MG 3 x 1 Fortaleza
29/10 – Internacional 1 x 1 América-MG
29/10 – Coritiba 1 x 2 Grêmio
29/10 – Bahia 1 x 0 Fluminense
29/10 – Cuiabá 0 x 2 Vasco
32ª rodada – 2º turno
01/11 – Fluminense 1 x 0 São Paulo
01/11 – Vasco 1 x 0 Botafogo
01/11 – Palmeiras 1 x 0 Athletico-PR
01/11 – Santos 0 x 0 Cuiabá
01/11 – Red Bull Bragantino 1 x 0 Corinthians
01/11 – Cruzeiro 1 x 2 Internacional
01/11 – Grêmio 1 x 0 Bahia
01/11 – Coritiba 0 x 1 Goiás
01/11 – Fortaleza 0 x 2 Flamengo
01/11 – América-MG 1 x 1 Atlético-MG
33ª rodada – 2º turno
05/11 – Flamengo 3 x 0 Palmeiras
05/11 – Botafogo 3 x 4 Grêmio
05/11 – São Paulo 1 x 0 Red Bull Bragantino
05/11 – Corinthians 1 x 1 Atlético-MG
05/11 – Goiás 0 x 1 Santos
05/11 – Cruzeiro 2 x 2 Vasco
05/11 – Internacional 0 x 0 Fluminense
05/11 – Athletico-PR 1 x 1 Fortaleza
05/11 – Bahia 0 x 3 Cuiabá
05/11 – América-MG 0 x 3 Coritiba
34ª rodada – 2º turno
12/11 – Flamengo 1 x 1 Fluminense
12/11 – Vasco 2 x 1 América-MG
12/11 – Palmeiras 3 x 0 Internacional
12/11 – Santos 0 x 0 São Paulo
12/11 – Red Bull Bragantino 2 x 2 Botafogo
12/11 – Atlético-MG 2 x 1 Goiás
12/11 – Grêmio 0 x 1 Corinthians
12/11 – Coritiba 1 x 0 Cruzeiro
12/11 – Bahia 1 x 1 Athletico-PR
12/11 – Cuiabá 2 x 1 Fortaleza
35ª rodada – 2º turno
22/11 – Fluminense 2 x 1 Coritiba
22/11 – Botafogo 1 x 1 Santos
22/11 – São Paulo 0 x 0 Cuiabá
22/11 – Corinthians 1 x 5 Bahia
22/11 – Goiás 0 x 1 Cruzeiro
22/11 – Atlético-MG 3 x 0 Grêmio
22/11 – Internacional 1 x 0 Red Bull Bragantino
22/11 – Athletico-PR 0 x 0 Vasco
22/11 – Fortaleza 2 x 2 Palmeiras
22/11 – América-MG 0 x 3 Flamengo
36ª rodada – 2º turno
26/11 – Flamengo 0 x 3 Atlético-MG
26/11 – Vasco 2 x 4 Corinthians
26/11 – Palmeiras 4 x 0 América-MG
26/11 – Santos 0 x 3 Fluminense
26/11 – Red Bull Bragantino 1 x 2 Fortaleza
26/11 – Cruzeiro 1 x 1 Athletico-PR
26/11 – Grêmio 2 x 1 Goiás
26/11 – Coritiba 1 x 1 Botafogo
26/11 – Bahia 0 x 1 São Paulo
26/11 – Cuiabá 0 x 2 Internacional
'''




l = list()

linhas = brasileirao.split('\n')



for linha in linhas:
    if 'rodada' in linha:
        continue
    info = linha.split(' – ')
    
    if len(info) == 2:
        
        data = info[0]
        jogo = info[1]


        info_data = data.split(' ')
        info_jogo = jogo.split(' x ')
        
        d = dict()
        
        d['data'] = info_data[0]
        d['mandante'] = info_jogo[0].rsplit(' ', 1)[0]
        d['visitante'] = info_jogo[1].split(' ', 1)[1]
        d['gols_mandante'] = info_jogo[0].rsplit(' ', 1)[1]
        d['gols_visitante'] = info_jogo[1].split(' ', 1)[0]
        
        l.append(d)
        
        
ultimos_jogos = pd.DataFrame(l)


```

```python
ultimos_jogos[['gols_mandante', 'gols_visitante']] = ultimos_jogos[['gols_mandante', 'gols_visitante']].astype(int)
```

```python
tabela_stats['média_gols_feitos_casa_ultimas_7_rod'] = ultimos_jogos.groupby('mandante')['gols_mandante'].mean().values
```

```python
tabela_stats['média_gols_feitos_fora_ultimas_7_rod'] = ultimos_jogos.groupby('visitante')['gols_visitante'].mean().values
```

```python
tabela_stats['gols_sofridos_fora_ultimas_7_rod'] = ultimos_jogos.groupby('visitante')['gols_mandante'].mean().values
```

```python
tabela_stats['gols_sofridos_casa_ultimas_7_rod'] = ultimos_jogos.groupby('mandante')['gols_visitante'].mean().values
```

```python
def calcular_pontuacao(linha):
    time_casa = linha["casa"]
    time_fora = linha["fora"]
    
    media_ponderada_casa = ((0.7 * tabela_stats.loc[tabela_stats["times"]==time_casa, "média_gols_feitos_casa_ultimas_7_rod"].iloc[0]) + 
                            (0.3 * tabela_stats.loc[tabela_stats["times"]==time_casa, "gols_feitos_casa"].iloc[0]))

    media_ponderada_fora = ((0.7 * tabela_stats.loc[tabela_stats["times"]==time_fora, 'média_gols_feitos_fora_ultimas_7_rod'].iloc[0]) + 
                            (0.3 * tabela_stats.loc[tabela_stats["times"]==time_fora, 'gols_feitos_fora'].iloc[0])
    )

    lambda_casa = media_ponderada_casa * tabela_stats.loc[tabela_stats["times"]==time_fora, "gols_sofridos_fora"].iloc[0]

    lambda_fora = media_ponderada_fora * tabela_stats.loc[tabela_stats["times"]==time_casa, "gols_sofridos_casa"].iloc[0]

    p_casa = 0
    p_empate = 0
    p_fora = 0

    for gols_time_casa in range(5):  # Considerando até 5 gols para cada time
        for gols_time_fora in range(5):
            prob_resultado = poisson.pmf(gols_time_casa, lambda_casa) * poisson.pmf(gols_time_fora, lambda_fora)
            if gols_time_casa > gols_time_fora:
                p_casa += prob_resultado
            elif gols_time_casa == gols_time_fora:
                p_empate += prob_resultado
            else:
                p_fora += prob_resultado
                
    ve_casa = p_casa * 3 + p_empate
    ve_fora = p_fora * 3 + p_empate
    linha["pontos_casa"] = ve_casa
    linha["pontos_fora"] = ve_fora
    return linha

```

```python
tabela_jogos_faltantes = tabela_jogos_faltantes.apply(calcular_pontuacao, axis=1)
display(tabela_jogos_faltantes)
```

```python
def ajustar_nome_time(linha):
    for nome in nomes_times:
        if nome in linha["Equipevde"]:
            return nome
    
tabela_classificacao["time"] = tabela_classificacao.apply(ajustar_nome_time, axis=1)
tabela_classificacao_atualizada = tabela_classificacao[["time", "Pts"]]
tabela_classificacao_atualizada["Pts"] = tabela_classificacao_atualizada["Pts"].astype(int)

display(tabela_classificacao_atualizada)
```

```python
tabela_pontuacao_casa = tabela_jogos_faltantes.groupby("casa").sum(numeric_only=True)[["pontos_casa"]]
tabela_pontuacao_fora = tabela_jogos_faltantes.groupby("fora").sum(numeric_only=True)[["pontos_fora"]]
display(tabela_pontuacao_casa)
display(tabela_pontuacao_fora)
```

```python
def atualizar_pontuacao_previsao(linha):
    time = linha["time"]
    pontuacao = int(linha["Pts"]) + float(tabela_pontuacao_casa.loc[time, "pontos_casa"]) + float(tabela_pontuacao_fora.loc[time, "pontos_fora"])
    return pontuacao


tabela_classificacao_atualizada["Pts"] = tabela_classificacao_atualizada.apply(atualizar_pontuacao_previsao, axis=1)
tabela_classificacao_atualizada = tabela_classificacao_atualizada.sort_values(by="Pts", ascending=False)
tabela_classificacao_atualizada = tabela_classificacao_atualizada.reset_index(drop=True)
tabela_classificacao_atualizada.index = tabela_classificacao_atualizada.index + 1
display(tabela_classificacao_atualizada)
```
